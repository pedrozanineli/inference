```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
files = os.listdir('emissions')
dfs_list = {}
for file in files:
    df = pd.read_csv(f'emissions/{file}')
    dfs_list[file[:-4]] = df
```

```python
columns = list(list(dfs_list.values())[0].columns)
```

```python
for column in columns[4:14]:

    plt.bar(list(dfs_list.keys()),[df[column].values[0] for df in dfs_list.values()])

    # plt.bar(['Mace','Orb'],[mace[column].values[0],orb[column].values[0]])
    plt.title(column)
    plt.show()
```
